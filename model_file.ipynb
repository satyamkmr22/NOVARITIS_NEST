```python
!pip install datasets
import torch
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset # Change the import to datasets
import pandas as pd
from torch.utils.data import Dataset as TorchDataset
```

```python
# Load BioBERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-v1.1")

# Function to tokenize textual columns
def tokenize_function(text_data):
    return tokenizer(text_data, padding='max_length', truncation=True, max_length=512)

# Tokenizing the textual columns
X_train_tokenized = X_train[textual_cols].apply(lambda row: ' '.join(row.astype(str)), axis=1)
X_test_tokenized = X_test[textual_cols].apply(lambda row: ' '.join(row.astype(str)), axis=1)

train_encodings = tokenizer(list(X_train_tokenized), padding=True, truncation=True, max_length=512)
test_encodings = tokenizer(list(X_test_tokenized), padding=True, truncation=True, max_length=512)
```

```python
# Add numerical columns to the tokenized text
class CustomDataset(TorchDataset):
    def __init__(self, encodings, numerical_data, labels):
        self.encodings = encodings
        self.numerical_data = numerical_data
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        # Access numerical features directly by index and ensure they are float32
        item['numerical_features'] = self.numerical_data[idx].type(torch.float32)  # Change to float32
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.float32) # Change to float32
        return item

    def __len__(self):
        return len(self.labels)

# Ensure numerical data is included
numerical_train = X_train[numerical_cols]
numerical_test = X_test[numerical_cols]

# Standardize numerical features (already done, assuming they're scaled)
# Ensure numerical data is float32
numerical_train = torch.tensor(numerical_train.values, dtype=torch.float32)  # Change to float32
numerical_test = torch.tensor(numerical_test.values, dtype=torch.float32)  # Change to float32

# Create the train and test datasets
train_dataset = CustomDataset(train_encodings, numerical_train, y_train.values)
test_dataset = CustomDataset(test_encodings, numerical_test, y_test.values)
```

```python
# Load BioBERT model
model = AutoModelForSequenceClassification.from_pretrained("dmis-lab/biobert-v1.1", num_labels=1)

# Define the training arguments
training_args = TrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=3,              # number of training epochs
    per_device_train_batch_size=8,   # batch size for training
    per_device_eval_batch_size=8,    # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
    evaluation_strategy="epoch",     # evaluate after each epoch
    save_strategy="epoch",           # save checkpoint after each epoch
    load_best_model_at_end=True      # load the best model when finished training
)
```

```python
# Create the Trainer
trainer = Trainer(
    model=model,                         # the instantiated 🤗 transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset            # evaluation dataset
)

# Train the model
trainer.train()
```

```python
trainer.evaluate()
```
